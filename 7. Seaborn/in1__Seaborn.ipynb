```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1=sns.load_dataset("tips")
df1
```

```python
sns.relplot(x='tip', y='total_bill', data=df1)
```

```python
#Hue is nothing but the differntiation between parameters on the basis of colors. 
sns.relplot(x='tip', y='total_bill', data=df1, hue='smoker')
```

```python
sns.relplot(x='tip', y='total_bill', data=df1, hue='day')
```

```python
#want to differentiate between smokers and non-smokers too.
sns.relplot(x='tip', y='total_bill', data=df1, hue='day', style='smoker')
```

```python
sns.relplot(x='tip', y='total_bill', data=df1, hue='day', style='smoker', size='size')
```

### Jointplot:

```python
sns.jointplot(data=df1, x='total_bill', y='tip') #scatter + histo
```

```python
try:
    sns.jointplot(data=df1, x='total_bill', y='tip', kind = 'line') #line + histo
except Exception as e:
    print(e)
```

```python
sns.jointplot(data=df1, x='total_bill', y='tip', kind = 'hist')
```

```python
sns.jointplot(data=df1, x='total_bill', y='tip', kind = 'reg') #facet-line
```

```python
sns.jointplot(data=df1, x='total_bill', y='tip', kind = 'resid')
```

### kde i.e. a non-parametric way to estimate the probability density function of a random variable.

```python
sns.jointplot(data=df1, x='total_bill', y='tip', kind = 'kde') #kde
```

```python
#Scatter Matrix
sns.pairplot(iris) # equivalent of pd.plotting.scattermatrix(df)
```

```python
'''
Plot data and regression model fits across a FacetGrid.
'''
sns.lmplot(data=df1, x='total_bill', y='tip') # Regressed plot even w/o using regression.
```
