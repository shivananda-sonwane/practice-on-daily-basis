## Task:

## fetch all the users data and display it in tabular form(DataFrame):

```python
#1
users_data={}
for i in j_data:
    for j in i["user"]:
        users_data[j] = []
users_data
#2
for i in range(len(j_data)):
    for j in j_data[i]['user']: #parsing each dictionary.
        x = j_data[i]['user'][j] #parsing each key of the sub dictionary.
        users_data[j].append(x)
users_data
```

```python
type(users_data)
```

```python
users_df=pd.DataFrame(users_data) 
users_df
```

```python
users_df.to_csv("users_data.csv", index=None)
```
