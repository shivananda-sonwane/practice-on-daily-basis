### Create a column having total price of items i.e. quantity * item_price:

```python
price = [float(ds1['item_price'][i][1:]) for i in range(len(ds1))]

#creating a column that does the needful
ds1['Amount_paid'] = price * ds1['quantity']

#adding $ before the total amount paid:

#creating a list that contains the total amount paid with $ sign.
dollars=[f"${str(ds1['Amount_paid'][i])}" for i in range(len(ds1))]

#implementing the column we desired.
ds1['Amount_paid'] = dollars
ds1
```

```python
#Creating a list that will contain the no. of times an item is bought.
timesbought = []

#filling in the values.
for i in ds1['items_offered'][ds1['items_offered'].isnull() == False]:
    timesbought.append(ds1[['quantity']][ds1['item_name']==i].sum())

#Converting the list into DataFrame datatype. 
timesbought = pd.DataFrame(timesbought)

#Creating a column we desired and assigning it the values we've had in the list. 
ds1[["No. of times an item is bought"]] = timesbought

ds1.head()
```

# DataSet 3: US Crime Rates (1960-2014)

```python
crime=pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv")
crime.head(10)
```

### Convert the type of the column Year to datetime64

```python
crime.dtypes
```

```python
crime.Vehicle_Theft.head()
#or
crime['Vehicle_Theft'].head()
```

```python
crime['Year'] = pd.to_datetime(crime['Year'], format='%Y') #Typecasting
crime.info()
crime.head()
crime
```

### Set the Year column as the index of the dataframe

```python
crime.index = crime.Year
crime = crime.drop(['Year'], axis=1)
crime.head()
```

```python
# OR
try:
    crime = crime.set_index('Year', drop = True)
except Exception as e:
    print(e)
crime.head()
```

### Group the Year by decades and sum the values accordingly:

```python
crime[::10] 
```

```python
#using resample to resample the data
crimes=crime.resample('10AS').sum()

#We oughta take the max population of the decade rather than the sum
#pretty much makes sense
pop = crime['Population'].resample('10AS').max()

#assigning the max population column to our new df crimes.
crimes['Population'] = pop

crimes
```
