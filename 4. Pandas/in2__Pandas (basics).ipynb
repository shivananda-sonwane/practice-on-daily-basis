### Data Serialization formats:

Data serialization is the process of **converting data objects present in complex data structures into a byte stream** for storage, transfer and distribution purposes on physical devices.

- ### Binary form of data is the format in which system stores our data by serializing where serializing is the conversion of the objects into byte-form.
- ### Pickle format stores our data in byte-from (binary format) by serializing it.
- ### And when we try to read that pickle file, it will present our data in the original from by de-serializing.
