### Fill the Na vals of each column with the average of the corrsoponding column accordingly.

```python
df2[i][[type(ele) != str for ele in df2[i]]].dropna()
```

```python
m = 0 #mean of numeric vals of one column at a time.
for i in df2.columns:
    
    m = df2[i][[type(ele) != str for ele in df2[i]]].dropna().mean()
    
    df2[i] = df2[i].fillna(m) #Here you are updating the columns of df2 and consequently df1. 
df2
```

### pd.merge():

```python
df1 = pd.DataFrame({'key': ['K0', 'K8', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
df1
```

```python
df2
```

### how = 'inner':

```python
pd.merge(df1, df2, on='key')
```

### left: just like SQL(Returns all records from the left table, and the matched records from the right table).

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='key', how = 'left')
```

### right: just like SQL(Returns all records from the right table, and the matched records from the left table).

```python
pd.merge(df1, df2, on='key', how = 'right')
```

### outer: just like SQL(Returns all records when there is a match in either left or right table).

```python
pd.merge(df1, df2, on='key', how = 'outer') #kinda union.
```

### .join(): Merges data based on indexes.

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', "A4"],
                     'B': ['B0', 'B1', 'B2', "B4"]},
                      index=['K0', 'K1', 'K2', 'K4']) 

df2 = pd.DataFrame({'C': ['C0', 'C2', 'C3', "C5"],
                    'D': ['D0', 'D2', 'D3', "D5"]},
                      index=['K0', 'K2', 'K3', 'K5'])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2) #All data from left and common part from right.
```

```python
df2.join(df1) #All data from df2 and common part from df1.
```
