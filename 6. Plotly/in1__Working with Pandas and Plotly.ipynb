### Histogram: Graph representing range of data vs no. of times the data has occured between that range i.e frequency.

```python
df1.plot.hist()
```

```python
df1['sepal_width']
```

```python
df1['sepal_width'].plot.hist()
```

```python
df1.plot.hist(stacked=True)
```

```python
#Stacked: Overlapping.
#Bins: Number of histogram bins to be used. 
df1.plot.hist(stacked = True, bins=100, figsize=(10,10), orientation='horizontal')
```

```python
df1.hist()
```

```python
df1.hist(figsize=(10,10), color=['r'], alpha=0.5)
```

### Scatter plot: Scatter of one parameter w respect to another.

```python
df1
```

```python
df1.plot(kind='scatter', x='sepal_width', y='sepal_length', color=['g'], alpha=0.5)
```

```python
df1.plot.scatter(x='sepal_width', y='sepal_length', color=['g'], alpha=0.5)
```

```python
#Color_scale: Tryna create a third dependency.
#Now here, petal_length will decide the color of our scattering points.
df1.plot.scatter(x='sepal_width', y='sepal_length', c='petal_length', alpha=0.5) 
```

```python
#A parameter scattering w.r.t to itself.
df1.plot.scatter(x='sepal_width', y='sepal_width', c='petal_length', alpha=0.5, label='scattering', s=180) 
```

```python
df1.plot.hexbin(x='sepal_width', y='sepal_length', alpha=0.5, gridsize=10) 
```
