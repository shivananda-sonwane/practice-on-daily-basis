**Note: Any point beyond the max point will be outlier and any point beyond the min point will be outlier.**

```python
d = [1,2,1,2,4,5,8,10,11,12,2,1,3,4,5,2,110,111]
df1 = pd.DataFrame([1,2,1,2,4,5,8,10,11,12,2,1,3,4,5,2,110,111])
df1.iplot(kind='box')
```

```python
import scipy.stats as stats
stats.iqr(d)
```

### To see the outliers:

**When reviewing a box plot, an outlier is defined as a data point that is located outside the whiskers of the box plot. For example, outside 1.5 times the interquartile range above the upper quartile and below the lower quartile (Q1 - 1.5 * IQR or Q3 + 1.5 * IQR).**

```python
sns.boxplot(data=d)
```

```python
#Also,
plt.boxplot(d)
plt.show()
```
