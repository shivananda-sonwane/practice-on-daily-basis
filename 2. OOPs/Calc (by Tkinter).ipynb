# Building a simple Calc:

```python
import os, logging as lg
from tkinter import *

log_file = "Calc.log"

if os.path.exists(log_file):
    os.remove(log_file)    
lg.basicConfig(filename = log_file, level = lg.INFO, format = "%(asctime)s %(message)s")

#adding a stream Handler to record the logs in console.
console_log = lg.StreamHandler()

#setting level to the Handler.
console_log.setLevel(lg.INFO)

#setting format to the Handler.
format = lg.Formatter("%(asctime)s %(level)s %(message)s")
console_log.setFormatter(format)

#Adding Handler to the logger.
lg.getLogger('').addHandler(console_log)
```

```python
'''
Creating a class Buttons to add the functionalities to the buttons.
'''
class Buttons:
    
    def __inint__(self):
        self.maths = "None"
        self.f_num = "" #to store the operand just when any operator button is clicked. 
        global e
        
    def b_click(self, number):
        try:
            current = e.get()
            e.delete(0, END)
            e.insert(0, current + str(number))
        except Exception as err:
            lg.error(err)
            
    def b_clear(self):
        try:
            e.delete(0, END)
        except Exception as err:
            lg.error(err)
            
    def _fetch(self): #protected function (ABSTRACTION).
        """
        fetching the first operand and clearing the entry box.
        """
        try:
            self.f_num = float(e.get())
            e.delete(0, END)
        except Exception as err:
            lg.error(err)
            
    def b_add(self):
        try:
            self._fetch()
            self.maths = "Add"
        except Exception as err:
            lg.error(err)
            
    def b_sub(self):
        try:
            self._fetch()
            self.maths = "Sub"
        except Exception as err:
            lg.error(err)

    def b_mul(self):
        try:
            self._fetch()
            self.maths = "Mul"
        except Exception as err:
            lg.error(err)

    def b_div(self):
        try:
            self._fetch()
            self.maths = "Div"  
        except Exception as err:
            lg.error(err)
            
    def equals(self):
        try:
            s_num = e.get()
            e.delete(0, END)

            if self.maths == "Add":
                e.insert(0, self.f_num + float(s_num))
                self.maths = "None"

            elif self.maths == "Sub":
                e.insert(0, self.f_num - float(s_num))
                self.maths = "None"

            elif self.maths == "Mul":
                e.insert(0, self.f_num * float(s_num))
                self.maths = "None"

            elif self.maths == "Div":
                e.insert(0, self.f_num / float(s_num)) 
                self.maths = "None"           

            elif self.maths == "None":
                e.insert(0, float(s_num))            

        except Exception as err:
            lg.error(err)
            
b = Buttons() #creating an object of the class Buttons.

root_calc = Tk() #creating our main window.
root_calc.title("SIMPLE CALC")

try:
    #input box which will prompt inputs from us and show the result of calculations done.
    e = Entry(root_calc, width = 60, borderwidth = 5)
    e.grid(row = 0, column = 0, columnspan = 4, padx = 10, pady = 10)


    #row1
    b_div = Button(root_calc, text = "/", width = 32, command = b.b_div)
    b_mul = Button(root_calc, text = "*", width = 15, command = b.b_mul)
    b_sub = Button(root_calc, text = "-", width = 15, command = b.b_sub)

    #row2
    b7 = Button(root_calc, text = 7, width = 15, command = lambda: b.b_click(7))
    b8 = Button(root_calc, text = 8, width = 15, command = lambda: b.b_click(8))
    b9 = Button(root_calc, text = 9, width = 15, command = lambda: b.b_click(9))

    #row3
    b4 = Button(root_calc, text = 4, width = 15, command = lambda: b.b_click(4))
    b5 = Button(root_calc, text = 5, width = 15, command = lambda: b.b_click(5))
    b6 = Button(root_calc, text = 6, width = 15, command = lambda: b.b_click(6))

    b_add = Button(root_calc, text = "+", width = 15, height = 3, command = b.b_add)

    #row4
    b1 = Button(root_calc, text = 1, width = 15, command = lambda: b.b_click(1))
    b2 = Button(root_calc, text = 2, width = 15, command = lambda: b.b_click(2))
    b3 = Button(root_calc, text = 3, width = 15, command = lambda: b.b_click(3))

    b_enter = Button(root_calc, text = "Enter", width = 15, height = 3, command = b.equals)

    #row5
    b0 = Button(root_calc, text = 0, width = 32, command = lambda: b.b_click(0))
    b_period = Button(root_calc, text = ".", width = 15, command = lambda: b.b_click('.'))

    #row6
    b_clear = Button(root_calc, text = "Clear", width = 65, command = b.b_clear)


    #arranging them:
    b_div.grid(row = 1, column = 0, columnspan = 2)
    b_mul.grid(row = 1, column = 2)
    b_sub.grid(row = 1, column = 3)


    b7.grid(row = 2, column = 0)
    b8.grid(row = 2, column = 1)
    b9.grid(row = 2, column = 2)

    b_add.grid(row = 2, rowspan = 2, column = 3)

    b4.grid(row = 3, column = 0)
    b5.grid(row = 3, column = 1)
    b6.grid(row = 3, column = 2)

    b1.grid(row = 4, column = 0)
    b2.grid(row = 4, column = 1)
    b3.grid(row = 4, column = 2)

    b_enter.grid(row = 4, rowspan = 2, column = 3)

    b0.grid(row = 5, column = 0, columnspan = 2)
    b_period.grid(row = 5, column = 2)

    b_clear.grid(row = 6, column = 0, columnspan = 4)
    
except Exception as err:
    lg.error(err)

finally:
    try:
        root_calc.mainloop()
    except Exception as err:
        lg.error(err)
```
