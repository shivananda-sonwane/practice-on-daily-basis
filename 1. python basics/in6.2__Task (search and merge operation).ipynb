# Task:

 - ## Creat a GUI app by tkinter. 
 - ## Two functionalities:
     - ### 1st: Scan; search for the input entered by scanning all directories or desired one and return all file names similar to the input.
     - ### 2nd: Merge; copy the data from the extension entered and dump into one single file having the same extension. 

```python
from tkinter import *
import os, logging as lg, docx

curDir = os.getcwd()

def mydir():
    global s_input, dirs, win1, prompt, l2, direc
    direc = s_input.get().capitalize()
            
    if direc in dirs:
        l2.config(text = f"Your desired directory {direc} will be scanned!\nNow, enter search input..")
        s_input.delete(0, END)
        return direc
    elif direc == "Search all" or direc == "Scan all":
        l2.config(text = "All directories will be scanned!\nNow, enter search input..")
        s_input.delete(0, END)
        return direc
    else:
        l2.config(text = "Enter an appropriate directory, my man!")
        s_input.delete(0, END)

def search_for():
    global direc, l4, fileRoot
    fileRoot = {}
    s = s_input.get() #searchable i.e. we'll search for it.
    s_input.delete(0, END)
    
    try:
        s_results.delete(1.0, END) #Clearing the prior search results if any before searching again.
        #print("generating search results...")
        l4.config(text = f"search results for '{s}'..")
        
        if direc == "Scan all" or direc == "Search all": #If all directories are to be searched.
            dirs1 = [i+"://" for i in dirs]
            for d in dirs1:
                for root, dirs, files in os.walk(d, topdown=True): 
                    for i in files:
                        if i.startswith(s) or s in i: 
                            """
                            if filenames start with s_input or s_inputs are present anywhere in between in filenames.
                            """
                            s_results.insert(END, i)
                            s_results.insert(END, '\n')
                            fileRoot[i] = root
        
        else: #if single directory is to be searched.
            for root, dirs, files in os.walk(f'{direc}://', topdown=True):
                    for i in files:
                        if i.startswith(s) or s in i: 
                            """
                            if filenames start with s_input or s_inputs are present anywhere in between in filenames.
                            """
                            s_results.insert(END, i)
                            s_results.insert(END, '\n')
                            fileRoot[i] = root
                    
    except Exception as e:
        print(e)
    else:
        #print("Search Results Generated!")
        pass
```

```python
def merge():
    global s_results, l7, fileRoot, content, curDir
    
    content = [i for i in fileRoot.keys()]
    curDir = os.getcwd() #current Directory.
    
    try:
        if ".txt" in content[0]:

            txt_file="txt(Merged).txt"
            if os.path.exists(txt_file):
                os.remove(txt_file)

            for i in content: #parsing each .txt file contained in the 'content'.
                os.chdir(fileRoot[i]) #going to the dir where the file is.
                
                with open(i, "r+") as f:
                    data = f.readlines()
                os.chdir(curDir) #coming back to our original directory.

                with open(txt_file, "a+") as g:
                    g.writelines(i)
                    g.write("\n")
                    g.write("\n")
                    g.writelines(data)
                    g.write("\n")
            g.close()
            print(f"Merged '.txt' is ready! You can check it now.")

        elif ".docx" in content[0]:
            docx_file = "docx(Merged).docx"
            if os.path.exists(docx_file):
                os.remove(docx_file)
                
            merged = docx.Document() #creating our merged .docx file
            for i in content:
                os.chdir(fileRoot[i])
                doc = docx.Document(i)
                
                os.chdir(curDir) #returning to our original dir.
                merged.add_heading(i)
                merged.add_paragraph("\n")
                for j in range(len(doc.paragraphs)):
                    merged.add_paragraph(doc.paragraphs[j].text)
                merged.add_paragraph("\n")
                
            merged.save(docx_file)
            print(f"Merged '.docx' is ready! You can check it now.")
            
        elif ".pdf" in content[0]:
            print(f"Merged '.pdf' is ready! You can check it now.")
            pass
        else:
            print(f"files of the extension '{content[0][-4:]}' coudn't be merged into a single file..")
            
    except Exception as e:
        print(e)
```

```python
win = Tk()
win.title("Merger")
#win.geometry("720x360" )
direc = "E"
dirs = ["C", "D", "E", "G"]

#row0
prompt = "Enter the desired directory from (C, D, E or G) \nOR\n Enter 'scan all' or 'search all' if you so much want so: "
l = Label(win, text = prompt)
l.grid(row = 0, column = 0)

#row1
s_input = Entry(win, borderwidth = 3, width = 100) #search input
s_input.grid(row = 1, column = 0)
chooseDir = Button(win, text = "select Dir", padx = 8, pady = 3, command = mydir).grid(row = 1, column = 1) #search button

#row2
l2 = Label(win, text = "")
l2.grid(row = 2, column = 0) #what directory has been chosen.

#row3
searchButton = Button(win, text = "search", padx = 10, pady = 5, command = search_for)
searchButton.grid(row = 3, column = 0)

#row4
l4 = Label(win)
l4.grid(row = 4, column = 0)

#row5
s_results = Text(win, borderwidth = 2) #search results
s_results.grid(row = 5, column = 0, columnspan=2)
#adding a vertical scroll bar.
vsbar = Scrollbar(win, orient='vertical')
vsbar.grid(row = 5, column = 2, sticky = 'e')
vsbar.config(command=s_results.yview)

#row6
b_merger = Button(win, text="Merge (.txt, .docx or .pdf)", padx=8, pady=3, command=merge).grid(row=6, column=0)

#row7
l7 = Label(win)
l7.grid(row = 7, column = 0)

win.mainloop()
```
