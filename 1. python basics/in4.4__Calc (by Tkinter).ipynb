# Building a Simple Calculator:

```python
from tkinter import *
global maths
maths = "None"
```

```python
def b_click(number):
    current = e.get()
    e.delete(0, END)
    e.insert(0, current + str(number))

def b_clear():
    e.delete(0, END)
    
def fetch(): 
    """
    fetching the first operand and clearing the entry box.
    """
    try:
        global f_num 
        f_num = int(e.get())
        e.delete(0, END)
        
    except Exception as err:
        print(err)
```

```python
def b_add():
    fetch()
    global maths
    maths = "Add"

def b_sub():
    fetch()
    global maths
    maths = "Sub"
    
def b_mul():
    fetch()
    global maths
    maths = "Mul"
    
def b_div():
    fetch()
    global maths
    maths = "Div"  
```

```python
def equals():
    try:
        global s_num  
        global maths
        s_num = int(e.get())
        e.delete(0, END)
        
        if maths == "Add":
            e.insert(0, f_num + int(s_num))
            maths = "None"
        
        elif maths == "Sub":
            e.insert(0, f_num - int(s_num))
            maths = "None"
            
        elif maths == "Mul":
            e.insert(0, f_num * int(s_num))
            maths = "None"
            
        elif maths == "Div":
            e.insert(0, f_num / int(s_num)) 
            maths = "None"           
            
        elif maths == "None":
            e.insert(0, s_num)            
        
    except Exception as err:
        print(err)
```

```python
root_calc = Tk() #our main window
root_calc.title("SIMPLE CALC")

#input box which will prompt inputs from us and show the result of calculations done.
e = Entry(root_calc, width = 60, borderwidth = 5)
e.grid(row = 0, column = 0, columnspan = 4, padx = 10, pady = 10)

#row1
b_div = Button(root_calc, text = "/", width = 32, command = b_div)
b_mul = Button(root_calc, text = "*", width = 15, command = b_mul)
b_sub = Button(root_calc, text = "-", width = 15, command = b_sub)

#row2
b7 = Button(root_calc, text = 7, width = 15, command = lambda: b_click(7))
b8 = Button(root_calc, text = 8, width = 15, command = lambda: b_click(8))
b9 = Button(root_calc, text = 9, width = 15, command = lambda: b_click(9))

#row3
b4 = Button(root_calc, text = 4, width = 15, command = lambda: b_click(4))
b5 = Button(root_calc, text = 5, width = 15, command = lambda: b_click(5))
b6 = Button(root_calc, text = 6, width = 15, command = lambda: b_click(6))

b_add = Button(root_calc, text = "+", width = 15, height = 3, command = b_add)

#row4
b1 = Button(root_calc, text = 1, width = 15, command = lambda: b_click(1))
b2 = Button(root_calc, text = 2, width = 15, command = lambda: b_click(2))
b3 = Button(root_calc, text = 3, width = 15, command = lambda: b_click(3))

b_enter = Button(root_calc, text = "Enter", width = 15, height = 3, command = equals)

#row5
b0 = Button(root_calc, text = 0, width = 32, command = lambda: b_click(0))
b_period = Button(root_calc, text = ".", width = 15, command = lambda: b_click('.'))

#row6
b_clear = Button(root_calc, text = "Clear", width = 65, command = b_clear)


#arranging them:
b_div.grid(row = 1, column = 0, columnspan = 2)
b_mul.grid(row = 1, column = 2)
b_sub.grid(row = 1, column = 3)


b7.grid(row = 2, column = 0)
b8.grid(row = 2, column = 1)
b9.grid(row = 2, column = 2)

b_add.grid(row = 2, rowspan = 2, column = 3)

b4.grid(row = 3, column = 0)
b5.grid(row = 3, column = 1)
b6.grid(row = 3, column = 2)

b1.grid(row = 4, column = 0)
b2.grid(row = 4, column = 1)
b3.grid(row = 4, column = 2)

b_enter.grid(row = 4, rowspan = 2, column = 3)

b0.grid(row = 5, column = 0, columnspan = 2)
b_period.grid(row = 5, column = 2)

b_clear.grid(row = 6, column = 0, columnspan = 4)

root_calc.mainloop()
```
