```python
'''
Create a function that takes input a number of mixed data types
and try to create a list of separate data based on individuals data types
and return multiple results.
'''
#method 1 (mine):
def ret_buck(*args):
    
    num_types = list(set([type(i) for i in args]))
    print(num_types)
    
    fin_bucket = []
    
    for i in num_types:
        bucket = []
        for j in args:
            if type(j) == i:
                bucket.append(j)
        fin_bucket.append(bucket)
    
    return fin_bucket
```

```python
ret_buck(34, 'suryansh', 22, "fyck it is?", 7.89, {'a': 'b', 'x': 'y'}, [34, 65, 87])
```

```python
#method 2:
def ret_buck2 (*args):
    
    d = {}
    
    for i in args:
        if type(i) not in d:
            d[type(i)] = [i]
        else:
            d[type(i)].append(i)
    
    return d
```

```python
ret_buck2(34, 'suryansh', 22, "fyck it is?", 7.89, {'a': 'b', 'x': 'y'}, [34, 65, 87])
```

```python
'''
Create a function which will be able to use args and kwargs
and it will be able to do all the list values concatenation
and return a list.
'''
def concat_list (*args, **kwargs):
    
    lst = []
    
    for i in args:
        if type(i) == list:
            lst += i
            
    for i in kwargs.values():
        if type(i) == list:
            lst += i
    
    return lst
```

```python
concat_list([1, 3, 4, 6, 7, 3, True, False, "fuck!"], a1 = 'b1', c1 = 'd1', jam = 'peanut butter', mobile = 'mobility', concat = [4, True, 357903, 8930574, 90, "Grover", 'damn!'])
```

```python
def ret_insideloop(n):
    lst = []
    for i in range(1, n+1):
        lst.append(i**3)
        
    return lst
```

```python
x = ret_insideloop(50000000)
x
```

```python
def yld_insideloop(n):
    print('[')
    for i in range(1, n+1):
        yield i**3
        if i != n:
            print(', ', end = '')
        if i == n:
            print(']')
```

```python
for i in yld_insideloop(600):
    print(i)
```
